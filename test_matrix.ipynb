```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.patches as patches

# Define two matrices
A = np.array([[1, 2, 3], [4, 5, 6]])  # 2x3
B = np.array([[7, 8], [9, 10], [11, 12]])  # 3x2
C = np.dot(A, B)  # Result: 2x2

# Set up the figure and axes
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(12, 3), gridspec_kw={'width_ratios': [3, 2, 3, 3]})
fig.suptitle("Matrix Multiplication Visualization: A (2x3) × B (3x2) = C (2x2)")

# Plot matrix A
ax1.matshow(A, cmap='Blues')
ax1.set_title('Matrix A')
for (i, j), val in np.ndenumerate(A):
    ax1.text(j, i, f'{val}', ha='center', va='center', color='black')

# Plot matrix B
ax2.matshow(B, cmap='Greens')
ax2.set_title('Matrix B')
for (i, j), val in np.ndenumerate(B):
    ax2.text(j, i, f'{val}', ha='center', va='center', color='black')

# Plot matrix C (result)
ax3.matshow(C, cmap='Reds')
ax3.set_title('Result C')
for (i, j), val in np.ndenumerate(C):
    ax3.text(j, i, f'{val}', ha='center', va='center', color='black')

# Text area for computation steps
ax4.axis('off')
ax4.set_xlim(0, 1)
ax4.set_ylim(0, 1)
comp_text = ax4.text(0.5, 0.5, '', ha='center', va='center', fontsize=12, wrap=True)

# Animation setup
m, n = C.shape  # Result matrix dimensions
steps = [(i, j) for i in range(m) for j in range(n)]  # All (i,j) positions in C

def update(frame):
    # Clear previous highlights
    for ax in [ax1, ax2, ax3]:
        for patch in ax.patches:
            patch.remove()
    
    # Get current (i,j) for result matrix C
    i, j = steps[frame]
    
    # Highlight row i in A
    rect_a = patches.Rectangle((-0.5, i-0.5), A.shape[1], 1, linewidth=2, edgecolor='red', facecolor='none')
    ax1.add_patch(rect_a)
    
    # Highlight column j in B
    rect_b = patches.Rectangle((j-0.5, -0.5), 1, B.shape[0], linewidth=2, edgecolor='red', facecolor='none')
    ax2.add_patch(rect_b)
    
    # Highlight position (i,j) in C
    rect_c = patches.Rectangle((j-0.5, i-0.5), 1, 1, linewidth=2, edgecolor='blue', facecolor='none')
    ax3.add_patch(rect_c)
    
    # Compute and display the dot product
    row_a = A[i, :]
    col_b = B[:, j]
    dot_product = np.sum(row_a * col_b)
    comp = f"C[{i},{j}] = {' + '.join([f'{a}×{b}' for a, b in zip(row_a, col_b)])} = {dot_product}"
    comp_text.set_text(comp)
    
    return [rect_a, rect_b, rect_c, comp_text]

# Create animation
ani = FuncAnimation(fig, update, frames=len(steps), interval=2000, blit=True)

# Display the plot
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np

def linear_least_squares(x, y):
    """
    Computes w and b using Linear Least Squares.
    Inputs: x (array of inputs), y (array of outputs)
    Outputs: w (slope), b (intercept)
    """
    # Create design matrix X with bias (ones) and x
    X = np.c_[np.ones(len(x)), x]
    # Solve X * theta = y for theta = [b, w]
    theta = np.linalg.lstsq(X, y, rcond=None)[0]
    b, w = theta
    return w, b

# Example usage with given data
x_data = np.array([0.5, 1.5, 3.0, 4.0, 5.0])
y_data = np.array([8.0, 6.0, 5.0, 2.0, 0.5])
w_lls, b_lls = linear_least_squares(x_data, y_data)
print(f"LLS: w = {w_lls:.4f}, b = {b_lls:.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_fitting_result(x, y, w, b, title="Fitting Result"):
    """
    Plots data points and the fitted line.
    Inputs: x, y (data), w, b (parameters), title (optional string)
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', label='Data Points')
    # Generate line points dynamically
    x_line = np.linspace(min(x) - 1, max(x) + 1, 100)  # Dynamic range
    y_line = w * x_line + b
    plt.plot(x_line, y_line, color='red', label=f'Fitted Line (w={w:.2f}, b={b:.2f})')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_fitting_result(x_data, y_data, w_lls, b_lls, title="LLS Fitting")
```

```python
import numpy as np

def lms_algorithm(x, y, eta, epochs, seed=42):
    """
    Runs LMS to find w and b.
    Inputs: x, y (data), eta (learning rate), epochs (int), seed (for random init/shuffle)
    Outputs: final w, b, w_history (list), b_history (list)
    """
    np.random.seed(seed)
    w = np.random.randn()
    b = np.random.randn()
    w_hist = [w]
    b_hist = [b]
    for epoch in range(epochs):
        indices = np.random.permutation(len(x))  # Shuffle for better convergence
        for ii in indices:
            y_pred = w * x[ii] + b
            e = y[ii] - y_pred
            w += eta * e * x[ii]
            b += eta * e
        w_hist.append(w)
        b_hist.append(b)
    return w, b, w_hist, b_hist

# Example usage
eta = 0.02
epochs = 100
w_lms, b_lms, w_hist, b_hist = lms_algorithm(x_data, y_data, eta, epochs)
print(f"LMS (η={eta}, epochs={epochs}): w = {w_lms:.4f}, b = {b_lms:.4f}")
```

```python
def plot_trajectories(w_hist, b_hist, eta, title="Weight Trajectories"):
    """
    Plots w and b over epochs.
    Inputs: w_hist, b_hist (lists), eta (for label), title (optional)
    """
    plt.figure(figsize=(8, 6))
    epochs_range = range(len(w_hist))
    plt.plot(epochs_range, w_hist, label='w (slope)', color='green')
    plt.plot(epochs_range, b_hist, label='b (intercept)', color='purple')
    plt.xlabel('Epochs (Learning Steps)')
    plt.ylabel('Parameter Value')
    plt.title(f"{title} (η={eta})")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_fitting_result(x_data, y_data, w_lms, b_lms, title="LMS Fitting (100 Epochs)")
plot_trajectories(w_hist, b_hist, eta, title="LMS Trajectories")
```

```python
def compare_lls_lms(x, y, eta, epochs, seed=42):
    """
    Computes and compares LLS and LMS.
    Inputs: x, y, eta, epochs, seed
    Outputs: Prints comparison
    """
    w_lls, b_lls = linear_least_squares(x, y)
    w_lms, b_lms, _, _ = lms_algorithm(x, y, eta, epochs, seed)
    print(f"LLS: w={w_lls:.4f}, b={b_lls:.4f}")
    print(f"LMS: w={w_lms:.4f}, b={b_lms:.4f}")
    print(f"Difference: Δw={abs(w_lls - w_lms):.4f}, Δb={abs(b_lls - b_lms):.4f}")

# Example usage
compare_lls_lms(x_data, y_data, 0.02, 100)
```

```python
def plot_comparison_fitting(x, y, w_lls, b_lls, w_lms, b_lms, title="LLS vs LMS Fitting"):
    """
    Plots data with both fitted lines.
    Inputs: x, y, w_lls, b_lls, w_lms, b_lms, title
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', label='Data Points')
    x_line = np.linspace(min(x) - 1, max(x) + 1, 100)
    plt.plot(x_line, w_lls * x_line + b_lls, color='red', label='LLS Fit')
    plt.plot(x_line, w_lms * x_line + b_lms, color='green', label='LMS Fit')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage (after computing)
plot_comparison_fitting(x_data, y_data, w_lls, b_lls, w_lms, b_lms)
```

```python
def study_learning_rates(x, y, etas, epochs, seed=42):
    """
    Runs LMS for multiple etas and prints results.
    Inputs: x, y, etas (list), epochs, seed
    Outputs: Dict of results
    """
    results = {}
    for eta in etas:
        w, b, w_hist, b_hist = lms_algorithm(x, y, eta, epochs, seed)
        results[eta] = {'w': w, 'b': b, 'w_hist': w_hist, 'b_hist': b_hist}
        print(f"η={eta}: w={w:.4f}, b={b:.4f}")
    return results

# Example usage
etas = [0.01, 0.05, 0.1]
results = study_learning_rates(x_data, y_data, etas, 100)
```

```python
def plot_multi_trajectories(results, title="Trajectories for Different η"):
    """
    Plots w and b histories for multiple etas.
    Inputs: results (dict from study), title
    """
    plt.figure(figsize=(10, 8))
    for eta, res in results.items():
        epochs_range = range(len(res['w_hist']))
        plt.plot(epochs_range, res['w_hist'], label=f'w (η={eta})')
        plt.plot(epochs_range, res['b_hist'], '--', label=f'b (η={eta})')
    plt.xlabel('Epochs')
    plt.ylabel('Parameter Value')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_multi_trajectories(results)

# For fittings (one per eta)
for eta, res in results.items():
    plot_fitting_result(x_data, y_data, res['w'], res['b'], title=f"LMS Fitting (η={eta})")
```

```python
import numpy as np

def leaky_lms_algorithm(x, d, eta, lambda_reg, epochs, seed=42):
    """
    Runs Leaky LMS to update w (and b as part of w if bias included).
    For single neuron: w is scalar, but code uses vectors for generality.
    Inputs: x (array of inputs), d (array of desired), eta, lambda_reg, epochs, seed
    Outputs: final w, w_history (list)
    """
    np.random.seed(seed)
    # Assume 1D for simplicity; add bias by augmenting x with 1s
    X_aug = np.c_[np.ones(len(x)), x]  # [1, x] for each sample
    w = np.random.randn(2)  # [b, w_slope]
    w_hist = [w.copy()]
    for epoch in range(epochs):
        indices = np.random.permutation(len(x))
        for ii in indices:
            pred = np.dot(w, X_aug[ii])
            e = d[ii] - pred
            # Leaky update: w = (1 - eta*lambda) w + eta * e * x_aug
            w = (1 - eta * lambda_reg) * w + eta * e * X_aug[ii]
        w_hist.append(w.copy())
    return w, w_hist

# Example usage (using previous data; treat as sequential)
x_data = np.array([0.5, 1.5, 3.0, 4.0, 5.0])
y_data = np.array([8.0, 6.0, 5.0, 2.0, 0.5])
eta = 0.02
lambda_reg = 0.1  # Example small lambda
epochs = 100
w_leaky, w_hist_leaky = leaky_lms_algorithm(x_data, y_data, eta, lambda_reg, epochs)
b_leaky, w_slope_leaky = w_leaky
print(f"Leaky LMS (η={eta}, λ={lambda_reg}): w={w_slope_leaky:.4f}, b={b_leaky:.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_leaky_fitting(x, d, w_vec, title="Leaky LMS Fitting"):
    """
    Plots data and fitted line (w_vec = [b, w])
    Inputs: x, d (data), w_vec, title
    """
    b, w = w_vec
    plt.figure(figsize=(8, 6))
    plt.scatter(x, d, color='blue', label='Data Points')
    x_line = np.linspace(min(x) - 1, max(x) + 1, 100)
    y_line = w * x_line + b
    plt.plot(x_line, y_line, color='orange', label=f'Leaky Fit (w={w:.2f}, b={b:.2f})')
    plt.xlabel('x')
    plt.ylabel('d')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_leaky_trajectories(w_hist, eta, lambda_reg, title="Leaky LMS Trajectories"):
    """
    Plots [b, w] over epochs from history list.
    Inputs: w_hist (list of arrays), eta, lambda_reg, title
    """
    plt.figure(figsize=(8, 6))
    epochs_range = range(len(w_hist))
    b_hist = [wh[0] for wh in w_hist]
    w_hist_vals = [wh[1] for wh in w_hist]
    plt.plot(epochs_range, w_hist_vals, label='w (slope)', color='green')
    plt.plot(epochs_range, b_hist, label='b (intercept)', color='purple')
    plt.xlabel('Epochs')
    plt.ylabel('Parameter Value')
    plt.title(f"{title} (η={eta}, λ={lambda_reg})")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_leaky_fitting(x_data, y_data, w_leaky)
plot_leaky_trajectories(w_hist_leaky, eta, lambda_reg)
```
